```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv("../data/taiwan_traffic_accident_a2_2021-1_to_2021-6.csv")
data2 = pd.read_csv("../data/taiwan_traffic_accident_a2_2021-7_to_2021-12.csv")

df1 = pd.DataFrame(data1)
df1 = df1[df1['發生地點'].notnull()]

df2 = pd.DataFrame(data2)
df2 = df2[df2['發生地點'].notnull()]

print(df1['發生時間'])
print(df2['發生時間'])

print(df2['經度'])
```

```python
list1 = [int(df1['發生時間'][i][11:13]) for i in range(len(df1.index))]
occ1  = [list1.count(i) for i in range(24)]

list2 = [int(df2['發生時間'][i][11:13]) for i in range(len(df2.index))]
occ2  = [list2.count(i) for i in range(24)]

total_occ = [x + y for x, y in zip(occ1, occ2)]

result_df = pd.DataFrame(data=total_occ)
result_df.plot.bar(xlabel='Hour of the day', ylabel='Number of accidents', title='Taiwan A2 Traffic Accident in 2021', legend=False)
```
